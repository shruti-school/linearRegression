**Q0.** Please answer the following questions in your own words.

1. What makes a model "linear"? "Linear" in what?

A model is "linear" if the relationship between the input and the output can be expressed as a linear combination of the input and other parameters. The correlation between the input and the output can be mapped by some sort of straight line on a coordinate plane.

2. How do you interpret the coefficient for a dummy/one-hot-encoded variable? (This is a trick question, and the trick involves how you handle the intercept of the model.)

The coefficient for a dummy or one-hot-encoded variable represents the change in the outcome when that specific category is present, relative to the baseline category, which is usually defined by the intercept term.

3. Can linear regression be used for classification? Explain why, or why not.

linear regression cannot be used for classification because it predicts continuous values rather than discrete class labels. This means that it predicts numbers rather than labeling items into a category. Models specifically designed for classification, like logistic regression, are better suited for these tasks.

4. What are signs that your linear model is over-fitting?

Your linear model is overfitting if it performs very well on training data but it performs bad on testing data or new data that it has never seen before. 

5. Clearly explain multi-colinearity using the two-stage least squares technique.

Multicollinearity is when two or more predictor variables are highly correlated in a regression model. This makes it hard to identify each indvidiual predictor’s effect. The two-stage least squares technique addresses this by using a variable—one that is uncorrelated with the error term. This separation allows for more accurate estimation of the predictor’s effect in the second-stage regression, reducing the impact of multicollinearity.

6. What are two ways to incorporate nonlinear relationships between your target/response/dependent/outcome variable $y$ and your features/control/response/independent variables $x$?

The first method is applying transformations to the features, such as taking the square, square root, or logarithm of x. The second way is adding polynomial terms or interaction terms to the model.

7. What is the interpretation of the intercept? A slope coefficient for a variable? The coefficient for a dummy/one-hot-encoded variable?

The intercept represents the  value of the outcome when all predictors are zero. A slope coefficient for a variable represents the expected change in the outcome variable for a one-unit increase in that predictor.
The coefficient for a dummy or one-hot-encoded variable represents the difference in the outcome between the group identified by the dummy variable and the baseline or reference group, defined by the intercept or other omitted category.

**Q4.** This is a math question to review the derivation of the OLS estimator (but only if you are into that kind of thing!). We are going to do it slightly differently from what we did in class, though. We will use a linear predictor and minimize the Sum of Squared Errors, just as in class. But, we are going to de-mean $X$ first, creating another variable $z_i = x_i - \bar{x}$ where
$$
\bar{x} = \dfrac{1}{N} \sum_{i=1}^N x_i,
$$
so the model is $\hat{y}_i = a + b z_i$ and the `SSE` is
$$
\text{SSE}(a,b) = \sum_{i=1}^N (y_i - a - bz_i)^2.
$$

  1. Take partial derivatives of the `SSE` with respect to $a$ and $b$. You should get

\begin{alignat*}{3}
\sum_{i=1}^N -2(y_i - a- bz_i) &=& 0 \\
\sum_{i=1}^N -2(y_i - a - bz_i)z_i &=& 0.
\end{alignat*}

  2. Solve for the solutions to the above equations. Big hint: $\bar{z} = 0$, since we subtracted the mean of $x$ from $x$ to get $z$. You should get

\begin{alignat*}{3}
a^* &=& \bar{y} \\
b^* &=& \dfrac{\sum_{i=1}^N(y_i - \bar{y})z_i}{\sum_{i=1}^N z_i^2}.
\end{alignat*}

  3. Substitute $z_i = x_i - \bar{x}$ back into the above equations. You should get
  
\begin{alignat*}{3}
a^* &=& \bar{y} \\
b^* &=& \dfrac{\sum_{i=1}^N(y_i - \bar{y})(x_i-\bar{x})}{\sum_{i=1}^N (x_i-\bar{x})^2},
\end{alignat*}

which can be written in terms of sample covariance and sample variance as:

\begin{alignat*}{3}
a^* &=& \bar{y} \\
b^* &=& \dfrac{\text{cov}(x,y)}{\text{var}(x)}.
\end{alignat*}

This is typically the preferred way of expressing the OLS coefficients.

4. When will $b^*$ be large or small, depending on the relationship between $x$ and $y$ and the amount of "noise"/variance in $x$? What does $a^*$ represent?
5. Suppose you have measurement error in $x$ which artificially inflates its variance (e.g. bad data cleaning). What happens to the $b^*$ coefficient? How will affect your ability to predict? (This phenomenon is called **attenuation**.)
